```python
class dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def bark(self):
        print(f'{self.name}が吠えました')
    def play(self, toy):
        print(f"{self.name}が{toy}で遊んでいます。")

pochi = dog('ポチ',3)
shiro = dog('シロ',5)
class new_dog(dog):
    def __init__(self,name,age,favorite):
        super().__init__(name,age)
        self.favorite = favorite
    def like(self):
        print(f'{self.name}は{self.favorite}が好きです')
tama = new_dog('たま',59,'ネズミ')


tama.like()
```
